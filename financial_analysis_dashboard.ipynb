```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("C:/Users/Salim Qureshi/Downloads/Finance Tracker_Workahop_Excel.xlsx")
df.head()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Month"] = df["Date"].dt.to_period("M")
```

```python
#grouping category spendind
Category_summary = df.groupby("Category")["Amount ($)"].sum()
print("Category spending:")
print(Category_summary)
```

```python
#monthwise spending
monthly_spending = df.groupby("Month")["Amount ($)"].sum()
print("Monthly spending:",monthly_spending)
```

```python
print("-------DASHBOARD--------")
total_spending = df['Amount ($)'].sum()
avg_spending = df['Amount ($)'].mean()
top_category = Category_summary.idxmax()

print(f"Total spending : {total_spending}")
print(f"Average spending: {avg_spending}")
print(f"Top category : {top_category}")
```

```python
# Category charts
Category_summary.plot(kind='bar')
plt.title("Category-wise spending")
plt.xlabel("Category")
plt.ylabel("Amount")
plt.xticks(rotation=45)
plt.show()
```

```python
# Monthly Trend
monthly_spending.plot(marker='o')
plt.title("Monthly spending Trend")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.grid()
plt.show()
```
